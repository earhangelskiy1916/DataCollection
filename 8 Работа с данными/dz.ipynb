```python
import pandas as pd
import numpy as np

# Загрузка датасета
df = pd.read_csv('train.csv')
df
```

# Предварительная обработка данных

```python
# a. Обработка отсутствующих значений
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
print("Отсутствующие значения в каждом столбце:\n", missing_values)
```

```python
# Заполнение отсутствующих значений
df['LotFrontage'].fillna(df['LotFrontage'].median(), inplace=True)
df['Alley'].fillna('None', inplace=True)
df['BsmtQual'].fillna('TA', inplace=True)
df['BsmtCond'].fillna('TA', inplace=True)
df['BsmtExposure'].fillna('No', inplace=True)
df['BsmtFinType1'].fillna('Unf', inplace=True)
df['BsmtFinType2'].fillna('Unf', inplace=True)
df['FireplaceQu'].fillna('Gd', inplace=True)
df['GarageType'].fillna('Attchd', inplace=True)
df['GarageFinish'].fillna('Unf', inplace=True)
df['GarageQual'].fillna('TA', inplace=True)
df['GarageCond'].fillna('TA', inplace=True)
df['PoolQC'].fillna('Ex', inplace=True)
df['Fence'].fillna('MnPrv', inplace=True)
df['MiscFeature'].fillna('None', inplace=True)
```

```python
# Удаление оставшихся строк с отсутствующими значениями
df.dropna(inplace=True)
```

```python
# b. Проверка на дубликаты
df.drop_duplicates(inplace=True)
```

```python
# c. Анализ и преобразование типов данных
df['MSSubClass'] = df['MSSubClass'].astype(str)
df['OverallCond'] = df['OverallCond'].astype(str)
df['YrSold'] = df['YrSold'].astype(str)
df['MoSold'] = df['MoSold'].astype(str)
```

# Разведочный анализ данных (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# a. Распределение SalePrice
plt.figure(figsize=(10,6))
sns.histplot(df['SalePrice'], kde=True)
plt.title('Распределение SalePrice')
plt.show()
```

```python
# b. Корреляционная матрица
# Выбираем только числовые столбцы
numeric_df = df.select_dtypes(include=[np.number])

# Строим корреляционную матрицу
corr_matrix = numeric_df.corr()

# Визуализируем корреляционную матрицу
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Корреляционная матрица')
plt.show()
```

```python
# c. Диаграммы рассеяния для важных характеристик
important_features = ['OverallQual', 'GrLivArea', 'GarageCars']
for feature in important_features:
    plt.figure(figsize=(10,6))
    sns.scatterplot(x=df[feature], y=df['SalePrice'])
    plt.title(f'{feature} vs SalePrice')
    plt.show()
```

# Проектирование признаков

```python
# a. Кодирование категориальных переменных
df = pd.get_dummies(df)
```

```python
# b. Создание новых признаков
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
```

```python
# Сохранение очищенного и преобразованного набора данных
df.to_csv('cleaned_house_prices.csv', index=False)
```
